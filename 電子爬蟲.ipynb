```python
import openpyxl
import time
import random
import requests
import datetime

wb = openpyxl.workbook

ws = wb.active

ws["A1"] = "機關名稱"
ws["B1"] = "標案名稱"
ws["C1"] = "連結"
ws["D1"] = "招標日期"
ws["E1"] = "截止日期"
ws["F1"] = "金額"


case_num=0


while case_num < 50:
    
    case_num += 1
    
    if soup.find_all("td",headers="addThId_j")[0].select("tr")[case_num] .select("td")[4].string  == str(int(str(datetime.date.today()).replace('-','/')[:4]) -1911)  + str(datetime.date.today()).replace('-','/')[4:] :
    
        res2=requests.get("https://web.pcc.gov.tw"  +  soup.find_all("td",headers="addThId_j")[0].select("tr")[case_num] .select("td")[3].a["href"])
        soup2 = BeautifulSoup(res2.text)

        a=soup.find_all("td",headers="addThId_j")[0].select("tr")[case_num]  .select("td")[2].string
        b=soup.find_all("td",headers="addThId_j")[0].select("tr")[case_num] .select("td")[3].div.text
        c="https://web.pcc.gov.tw"  +  soup.find_all("td",headers="addThId_j")[0].select("tr")[case_num] .select("td")[3].a["href"]
        d=soup.find_all("td",headers="addThId_j")[0].select("tr")[case_num]  .select("td")[4].text
        e=soup.find_all("td",headers="addThId_j")[0].select("tr")[case_num] .select("td")[6].text
        
        tr = 45
        while tr < 67:
            tr += 1
            if soup2.select("tr")[tr].th != None:
                if soup2.select("tr")[tr].th.text.strip() == "預算金額":
                    f=soup2.select("tr")[tr].td.text.strip()
        print("以讀取",case_num)
        ws.append([a,b,c,d,e,f])
        print("-------------------------------------------------------------------------------")
        time.sleep(random.randint(8,13))

    wb.save("招標清單"+str(datetime.date.today())+".xlsx")
print("抓取結束")
```
